```python
import sys
import os
sys.path.append(os.path.abspath(os.path.join(os.getcwd(), '..')))

from src.preprocessor import Preprocessor
import pandas as pd

# Load raw data
df = pd.read_csv('../data/raw/complaints.csv')

# Process data
products_list = ["Credit card", "Personal loan", "Savings account", "Money transfers"]

preprocessor = Preprocessor(df)
df_processed = (
    preprocessor
    .filter_products(products_list)
    .remove_empty_narratives()
    .preprocess_narratives()
    .get_df()
)

# Save cleaned data
df_processed.to_csv('../data/processed/complaints_clean.csv', index=False)
```
