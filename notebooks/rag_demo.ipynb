```python
import pandas as pd

# 1. Representative evaluation questions
questions = [
    "Why are people unhappy with credit cards?",
    "Why are some people denied credit cards even with good credit?",
    "What problems do customers report about interest rates?",
    "How do credit card companies push people into debt?",
    "What financial unfairness do consumers complain about?"
]

results = []

# 2. Run RAG for each question
for q in questions:
    answer, sources = rag.ask(q)

    # Take top 2 sources
    top_sources = sources[:2]

    source_text = "\n".join([
        f"{s['product']} (ID {s['complaint_id']}): {s['text'][:120]}..."
        for s in top_sources
    ])

    results.append({
        "Question": q,
        "Generated Answer": answer,
        "Retrieved Sources": source_text,
        "Quality Score (1–5)": "",  # You will fill this manually
        "Comments / Analysis": ""   # You will fill this manually
    })

# 3. Convert to DataFrame
df_eval = pd.DataFrame(results)

# 4. Display table
df_eval
```
