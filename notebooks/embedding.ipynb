```python

import sys
import os
import pandas as pd
sys.path.append(os.path.abspath(os.path.join(os.getcwd(), '..')))
from src.embedding import Embedder

# Load cleaned data
df = pd.read_csv("../data/processed/complaints_clean.csv")

# Initialize embedder
embedder = Embedder(df)

# Step 1: Stratified sample ~12k complaints
embedder.stratified_sample(n_samples=12000)

# Step 2: Chunk long narratives
embedder.chunk_texts(chunk_size=500, chunk_overlap=50)

# Step 3: Generate embeddings
embedder.embed_chunks()

# Step 4: Create and save FAISS vector store
embedder.create_faiss_index(index_path="../vector_store/faiss_index")
```
